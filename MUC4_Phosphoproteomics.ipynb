# The trans effects of MUC4 mutation on known interacting proteins (proteomic analysis)
Next we wanted to understand how TP53 mutation impacted proteins in p53's interaction network. We discovered many proteins whose abundance significantly changed in the presence of TP53 mutation, but none of them were exclusively due to hotspot mutations.

### Standard imports, including CPTAC package

```python
import pandas as pd
import numpy as np
import re
import scipy.stats
import statsmodels.stats.multitest
import matplotlib.pyplot as plt
import seaborn as sns
import CPTAC.Ovarian as CPTAC
```

## Set up initial dataframes and variables

```python
gene='MUC4'
protein = 'ERBB2'
somatic = CPTAC.get_somatic_mutations()
proteomics = CPTAC.get_proteomics()
phos = CPTAC.get_phosphoproteomics()
transcriptomics = CPTAC.get_transcriptomics()
```

```python
protList = ['AKT', 'PIK3CA', 'ERBB2', 'HER2']
```

## MUC4 Mutation Alone

```python
mutation_data = CPTAC.compare_mutations(phos, protList[0], gene)[['Mutation']]

for protein in protList:
    prot_phos = CPTAC.compare_mutations(phos, protein, gene)
    prot_phos.drop(['Mutation', 'patient_key', 'Sample_Status'], axis=1, inplace=True)
    prot_phos = prot_phos.groupby(prot_phos.columns, axis=1).agg(np.nanmean, axis=1)
    for site in prot_phos.columns:
        site_phos = pd.concat([mutation_data, prot_phos[site]], axis=1, sort=False).dropna(axis=0)
        site_phos = site_phos[~site_phos.Mutation.str.contains('Wildtype_Normal')]
        wt_muc4 = site_phos.loc[site_phos['Mutation'] == 'Wildtype_Tumor']
        mutated_muc4 = site_phos.loc[site_phos['Mutation'] == 'Missense_Mutation']
        if len(wt_muc4[site]) < 2 or len(mutated_muc4[site]) < 2: continue
        ttest = scipy.stats.ttest_ind(wt_muc4[site], mutated_muc4[site])
        if(ttest[1] < 0.05):
            print(site + ': ' + str(ttest[1]))     
```

## Combination of mutations in MUC4 and MUC17

## Find patients that have both mutations

```python
second_gene = 'MUC17'
muc4_mut = CPTAC.compare_mutations(proteomics, 'CDH2', gene)
muc16_mut = CPTAC.compare_mutations(proteomics, 'CDH2', second_gene)
new_mutation = pd.DataFrame()
new_mutation[gene + ' Mutation'] = muc4_mut['Mutation']
new_mutation[second_gene + ' Mutation'] = muc16_mut['Mutation']
new_mutation['CDH2'] = muc4_mut['CDH2']
new_mutation = new_mutation.loc[new_mutation[gene + ' Mutation'] != 'Wildtype_Normal']
new_mutation = new_mutation.loc[new_mutation[gene + ' Mutation'] != 'Wildtype_Tumor'].loc[new_mutation[second_gene + ' Mutation'] != 'Wildtype_Tumor']
patients_both_mut_17 = new_mutation.index
patients_both_mut_17
```

```python
mutation_data = CPTAC.compare_mutations(phos, protList[0], gene)[['Mutation']]

for protein in protList:
    prot_phos = CPTAC.compare_mutations(phos, protein, gene)
    prot_phos.drop(['Mutation', 'patient_key', 'Sample_Status'], axis=1, inplace=True)
    prot_phos = prot_phos.groupby(prot_phos.columns, axis=1).agg(np.nanmean, axis=1)
    for site in prot_phos.columns:
        site_phos = pd.concat([mutation_data, prot_phos[site]], axis=1, sort=False).dropna(axis=0)
        site_phos = site_phos[~site_phos.Mutation.str.contains('Wildtype_Normal')]
        if np.isin(patients_both_mut_17, site_phos.index).any() == False: continue
        site_both_mut = patients_both_mut_17[np.isin(patients_both_mut_17, site_phos.index)]
        site_phos['Concurrent_Mutations'] = 'Single'
        site_phos.loc[site_both_mut, 'Concurrent_Mutations'] = 'Both'
        site_phos.loc[site_phos['Mutation'] == 'Wildtype_Tumor', 'Concurrent_Mutations'] = 'Wildtype'
        wt_muc4 = site_phos.loc[site_phos['Concurrent_Mutations'] == 'Wildtype']
        mutated_muc4 = site_phos.loc[site_phos['Concurrent_Mutations'] == 'Both']
        if len(wt_muc4[site]) < 2 or len(mutated_muc4[site]) < 2: continue
        ttest = scipy.stats.ttest_ind(wt_muc4[site], mutated_muc4[site])
        if(ttest[1] < 0.05):
            print(site + ': ' + str(ttest[1]))     
```
