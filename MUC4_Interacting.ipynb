# The trans effects of MUC4 mutation on known interacting proteins (proteomic analysis)
Next we wanted to understand how TP53 mutation impacted proteins in p53's interaction network. We discovered many proteins whose abundance significantly changed in the presence of TP53 mutation, but none of them were exclusively due to hotspot mutations.

### Standard imports, including CPTAC package

```python
import pandas as pd
import numpy as np
import re
import scipy.stats
import statsmodels.stats.multitest
import matplotlib.pyplot as plt
import seaborn as sns
import CPTAC.Ovarian as CPTAC
```

## Set up initial dataframes and variables

```python
gene='MUC4'
somatic = CPTAC.get_somatic_mutations()
proteomics = CPTAC.get_proteomics()
phos = CPTAC.get_phosphoproteomics()
transcriptomics = CPTAC.get_transcriptomics()
```

## Find the most commonly mutated proteins in our cohort
Note: Often a single patient will have more than one mutation in the same protein, so we need to account for that

```python
indx_somatic = somatic.index
somatic['index_copy'] = indx_somatic
mutation_frequency = somatic.drop_duplicates(subset=['index_copy', 'Gene'])
mutation_frequency = mutation_frequency.groupby('Gene').count().sort_values(by='Mutation', ascending=False)
print(mutation_frequency['Mutation'].head(20))
```

### List of proteins in MUC4's interaction network (according to Uniprot and String)
We compiled this list in order to narrow down the number of proteins we were going to test. We did this to avoid missing potential results due to needing to correct for testing thousands of proteins at once.
String interactions: https://string-db.org/cgi/network.pl?taskId=OTIMuBPCUH44
Lists up-to-date as of March 29th, 2019

```python
# Create lists of interacting proteins from both sources
protList = ['MET', 'MUC5AC', 'MUC1', 'MUC4', 'MUC5B', 'MUC15', 'MUC13',
       'MUC12', 'MUC7', 'STAT3', 'GRB2', 'MUC20', 'MUC6', 'GALNT4',
       'MUC16', 'ST3GAL1', 'GALNT11', 'GALNT13', 'MUC21', 'C1GALT1',
       'MUC3A', 'B4GALT2', 'GCNT1', 'B4GALT1', 'B3GNT6', 'GCNT3',
       'ST3GAL2', 'GALNTL6', 'GALNT6', 'ST3GAL4', 'B4GALT4', 'B3GNT3',
       'GCNT4', 'GCNT7', 'B4GALT3', 'B3GNT5', 'GALNTL5', 'GALNT12',
       'MUC17', 'ST6GALNAC4', 'ST6GALNAC3', 'GALNT14', 'GALNT15', 'MUCL1',
       'GALNT2', 'C1GALT1C1', 'GALNT3', 'GALNT9', 'GALNT16', 'GALNT7',
       'GALNT10', 'GALNT5', 'GALNT8', 'ST3GAL3', 'GALNT1', 'GALNT18', 'CDH2']
#Combine lists and remove duplicates
print('Number of proteins to test: ' + str(len(protList)) + '\n\n' + 'Protein List: \n')
print(protList)
```

### Adjust significance cutoff using a Bonferroni correction

```python
pcutoff = 0.05 / len(proteomics.columns)
```

### All mutations vs. cancer wildtype

## Find patients that have both mutations

```python
second_gene = 'MUC17'
muc4_mut = CPTAC.compare_mutations(proteomics, 'CDH2', gene)
muc16_mut = CPTAC.compare_mutations(proteomics, 'CDH2', second_gene)
new_mutation = pd.DataFrame()
new_mutation[gene + ' Mutation'] = muc4_mut['Mutation']
new_mutation[second_gene + ' Mutation'] = muc16_mut['Mutation']
new_mutation['CDH2'] = muc4_mut['CDH2']
new_mutation = new_mutation.loc[new_mutation['MUC4 Mutation'] != 'Wildtype_Normal']
new_mutation = new_mutation.loc[new_mutation[gene + ' Mutation'] != 'Wildtype_Tumor'].loc[new_mutation[second_gene + ' Mutation'] != 'Wildtype_Tumor']
patients = new_mutation.index
patients
```

## MUC17 results

```python
significant_all_mut_wt17 = {}

# Go through every protein in our proteomics dataframe
for protein in proteomics.columns:
    if (protein != 'patient_key'):
        muc4_mut_prot = CPTAC.compare_mutations(proteomics, protein, gene)
        muc4_mut_prot['Concurrent_Mutations'] = 'Single'
        muc4_mut_prot.loc[patients, 'Concurrent_Mutations'] = 'Both'
        muc4_mut_prot.loc[muc4_mut_prot['Mutation'] == 'Wildtype_Tumor', 'Concurrent_Mutations'] = 'Wildtype'
        muc4_mut_prot = muc4_mut_prot.dropna(axis = 0)
        wt_muc4 = muc4_mut_prot.loc[muc4_mut_prot['Concurrent_Mutations'] == 'Single']
        mutated_muc4 = muc4_mut_prot.loc[muc4_mut_prot['Concurrent_Mutations'] == 'Both']
        ttest = scipy.stats.ttest_ind(wt_muc4[protein], mutated_muc4[protein])
        if len(wt_muc4[protein]) < 2 or len(mutated_muc4[protein]) < 2: continue
        if len(wt_muc4.columns) > 5:
            if ttest[1][0] <= pcutoff:
                print("Found significant result! " + protein)
                print(ttest[1])
                significant_all_mut_wt[protein] = ttest[1]
        else:
            if ttest[1] <= pcutoff:
                print("Found significant result! " + protein)
                print(ttest[1])
                significant_all_mut_wt[protein] = ttest[1]
    
print('Proteins with significant changes in abundance in the presence of MUC4 mutation (n=' + str(len(significant_all_mut_wt)) + '):\n')
significant_all_mut_wt17
```

## MUC16 results

```python
significant_all_mut_wt = {}

# Go through every protein in our proteomics dataframe
for protein in proteomics.columns:
    if (protein != 'patient_key'):
        muc4_mut_prot = CPTAC.compare_mutations(proteomics, protein, gene)
        muc4_mut_prot['Concurrent_Mutations'] = 'Single'
        muc4_mut_prot.loc[patients, 'Concurrent_Mutations'] = 'Both'
        muc4_mut_prot.loc[muc4_mut_prot['Mutation'] == 'Wildtype_Tumor', 'Concurrent_Mutations'] = 'Wildtype'
        muc4_mut_prot = muc4_mut_prot.dropna(axis = 0)
        wt_muc4 = muc4_mut_prot.loc[muc4_mut_prot['Concurrent_Mutations'] == 'Single']
        mutated_muc4 = muc4_mut_prot.loc[muc4_mut_prot['Concurrent_Mutations'] == 'Both']
        ttest = scipy.stats.ttest_ind(wt_muc4[protein], mutated_muc4[protein])
        if len(wt_muc4[protein]) < 2 or len(mutated_muc4[protein]) < 2: continue
        if len(wt_muc4.columns) > 5:
            if ttest[1][0] <= 0.005:
                print("Found significant result! " + protein)
                print(ttest[1])
                significant_all_mut_wt[protein] = ttest[1]
        else:
            if ttest[1] <= 0.005:
                print("Found significant result! " + protein)
                print(ttest[1])
                significant_all_mut_wt[protein] = ttest[1]
    
print('Proteins with significant changes in abundance in the presence of MUC4 mutation (n=' + str(len(significant_all_mut_wt)) + '):\n')
significant_all_mut_wt
```

## Exploratory plot

```python
proteins_to_plot = []
for k, v in significant_all_mut_wt.items():
    proteins_to_plot.append(k)

proteins_to_plot = ['AHCY', 'CHD3', 'ZNF98', 'SHMT2', 'EFTUD2', 'USP39', 'PSMC2', 'PSMD1', 'TRIM21', 'KCTD12']

plotdf = CPTAC.compare_mutations(proteomics, gene)
plotdf = plotdf.loc[plotdf['Sample_Status'] == 'Tumor'][['Mutation']]
for protein in proteins_to_plot:
    proteindf = CPTAC.compare_mutations(proteomics, protein, gene)[[protein]]
    plotdf = plotdf.join(proteindf)

plotdf['Concurrent_Mutations'] = 'Single'
plotdf.loc[patients, 'Concurrent_Mutations'] = 'Both'
plotdf.loc[plotdf['Mutation'] == 'Wildtype_Tumor', 'Concurrent_Mutations'] = 'Wildtype'
plotdf = pd.melt(plotdf, id_vars = ['Mutation', 'Concurrent_Mutations'], var_name = 'Protein', value_name = 'Proteomics')
plotdf = plotdf.dropna(axis = 0)

a4_dims = (23.4, 16.54)
fig, ax = plt.subplots(figsize=a4_dims)

#Create the plot
boxplt = sns.boxplot(x='Protein', y='Proteomics', hue='Concurrent_Mutations', hue_order = ['Wildtype', 'Single', 'Both'], data=plotdf)
boxplt = sns.stripplot(data=plotdf, x='Protein', y='Proteomics', hue='Concurrent_Mutations', hue_order = ['Wildtype', 'Single', 'Both'], dodge=True, jitter=True, color='.3')

#Add styling
boxplt.set_xlabel('')
boxplt.set_ylabel('Protein Expression', fontsize='40')
boxplt.tick_params(labelsize='15')

#Adjust legend
handles, labels = ax.get_legend_handles_labels()
plt.legend(handles[0:3], ['Wildtype', 'Single', 'Both'], title='Concurrent Mutations', fontsize='20', frameon=False)
boxplt.get_legend().get_title().set_fontsize('20')
boxplt.get_legend().set_bbox_to_anchor((1, 1, 0, 0))
```
